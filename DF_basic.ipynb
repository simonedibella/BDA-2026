# Pandas DataFrame Intro

This notebook demonstrates how to work with a Pandas DataFrame.
We will:

- Load data from `data/symbol_info_3-25.csv` file, randomly sampling 25 rows, and selecting the desired columns.
- Display general information about the DataFrame.
- Show the content by position and by column
- Sort the DataFrame by `Market_cap`.
- Add a column with the percentage of each `Market_cap` relative to the total market cap.
- Grouping and aggregation
- Advanced Sorting and indexing
- Filtering
- Operations on non-numerical columns

## Loading and Sampling Data

We start by loading a CSV file into a DataFrame, then randomly selecting 25 rows and keeping only the columns we need: `symbol`, `market_cap`, and `sector`.

````python
# Import the pandas library (the main tool for working with tabular data in Python)
import pandas as pd

# Load all data from the CSV file into a DataFrame
df = pd.read_csv('data/symbol_info_3-25.csv')

# Randomly pick 25 rows from the full dataset
# random_state=42 ensures we always get the same 25 rows (for reproducibility)
df_sample = df.sample(25, random_state=77) # last three digits of my ID: 077

# Keep only the 3 columns we need
df_sample = df_sample[['symbol', 'market_cap', 'sector']]

# Display the result
print('--- Initial Sampled DataFrame ---')
print(df_sample)
````

## Exploring the DataFrame

Before working with data, it is important to understand its structure. We use:

- **`.info()`** — shows column names, data types, and non-null counts
- **`.describe()`** — gives summary statistics for numerical columns (mean, std, min, max, quartiles)
- **`.describe(include='str')`** — gives summary statistics for text columns (count, unique values, most frequent)

````python
# .info() prints a summary: column names, types, and how many non-null values
print('--- DataFrame Info ---')
df_sample.info()

# .describe() gives statistics for numerical columns (count, mean, std, min, max, quartiles)
print('\n--- DataFrame Description (numerical columns) ---')
print(df_sample.describe())

# .describe(include='str') gives statistics for text columns (count, unique values, most frequent)
print('\n--- DataFrame Description (text columns) ---')
print(df_sample.describe(include='object'))
````

## Accessing Rows by Position with `.iloc[]`

`.iloc[]` lets you select rows by their **position** (like indexing a Python list). Position `0` is the first row, `-1` is the last, `-2` is the second-to-last, and so on.

````python
# .iloc[] lets you access rows by their POSITION (like a list index)

# First row (position 0)
print('--- First Element (iloc[0]) ---')
print(df_sample.iloc[0])

# Access a specific column value from the first row
print('\n--- Market Cap of the first Element ---')
print(df_sample.iloc[0]['market_cap'])

# Last row (position -1, just like Python lists)
print('\n--- Last Element (iloc[-1]) ---')
print(df_sample.iloc[-1])

# Second-to-last row
print('\n--- Second-to-Last Element (iloc[-2]) ---')
print(df_sample.iloc[-2])
````

## Sorting a DataFrame

We can sort the rows of a DataFrame by any column using `.sort_values()`. Here we sort by `market_cap` in descending order (largest first).

````python
# Sort the DataFrame by Market_cap
df_sorted = df_sample.sort_values(by='market_cap', ascending=False)
print('--- DataFrame Sorted by Market_cap ---')
print(df_sorted)
````

## Adding a Computed Column

We can create a new column based on existing data. Here we compute the **percentage** that each company's market cap represents out of the total market cap of all 25 companies.

````python
# --- Add a new column: percentage of total market cap ---

# First, compute the total market cap across all 25 companies
total_market_cap = df_sample['market_cap'].sum()

# Divide each company's market_cap by the total and multiply by 100 to get a percentage
df_sample['market_cap_%'] = (df_sample['market_cap'] / total_market_cap) * 100

print('--- DataFrame with Market Cap Percentage ---')
print(df_sample)

# --- Alternative ways to do the same thing (commented out for reference) ---

# Using apply() with a lambda (a one-line anonymous function):
# df_sample['market_cap_%'] = df_sample['market_cap'].apply(lambda x: (x / total_market_cap) * 100)

# Using apply() with a named function:
# def compute_percentage(value):
#     return (value / total_market_cap) * 100
# df_sample['market_cap_%'] = df_sample['market_cap'].apply(compute_percentage)
````

````python
# Computing difference of share ratio from global average (mean) share ratio
mean = df_sample['market_cap_%'].mean()
df_sample['diff_from_mean'] = df_sample['market_cap_%'] - mean
df_sample 
````

## Grouping and Aggregation

Sometimes we want to **summarize data by category**. For example: *"What is the total market cap for each sector?"*

The idea is simple — split the data into groups (one per sector), then compute a summary for each group.

We do this in three steps:

1. **Group** the rows by a column (e.g., `sector`) using `groupby()`
2. **Pick** the column we want to summarize (e.g., `market_cap`)
3. **Apply** one or more summary functions (`sum`, `mean`, `count`)

````python
# Step 1: Group the rows by 'sector'
grouped = df_sample.groupby('sector')
print('Step 1 — type:', type(grouped).__name__)
print('Number of groups:', grouped.ngroups)
print('Group sizes:')
print(grouped.size())
print()

# Let's peek inside one group to see what it looks like
print('--- Example: the "Technology" group ---')
print(grouped.get_group('Technology'))
print()

# Step 2: Select the 'market_cap' column from each group
market_cap_by_sector = grouped['market_cap']
print('Step 2 — now we only keep the market_cap values per group:')
for name, values in market_cap_by_sector:
    print(f'  {name}: {values.tolist()}')
print()

# Step 3: Apply summary functions — sum, mean, and count
df_grouped = market_cap_by_sector.agg(['sum', 'mean', 'count'])
print('Step 3 — after agg():')
print(df_grouped)
print()
# Notice: 'sector' is now the index, not a regular column

# reset_index() turns 'sector' back into a regular column
df_grouped = df_grouped.reset_index()
print('After reset_index():')
print(df_grouped)

# NOTE: The 4 steps above can also be written in one line:
# df_grouped = df_sample.groupby('sector')['market_cap'].agg(['sum', 'mean', 'count']).reset_index()
````

## Advanced Sorting and Indexing

By default, a DataFrame uses row numbers (0, 1, 2, ...) as its index. But we can change the index to something more meaningful:

- **`set_index('symbol')`** — use the stock symbol as the row label, so we can look up rows by name (e.g., `df.loc['PYPL']`)
- **Multi-index** — use *two* columns as a combined index (e.g., `sector` + `symbol`), which organizes the data hierarchically
- **Sorting by multiple columns** — for example, sort by `sector` alphabetically, and within each sector sort by `market_cap` from largest to smallest

````python
# --- Using set_index() ---
# Replace the default numeric index with 'symbol', so each row is labeled by its stock ticker
df_indexed = df_sample.set_index('symbol')
print('--- DataFrame with symbol as index ---')
print(df_indexed)

# --- Multi-Index ---
# Use TWO columns as the index: first 'sector', then 'symbol'
# This creates a hierarchical structure: sector > symbol
df_multi_indexed = df_sample.set_index(['sector', 'symbol'])
print('\n--- DataFrame with Multi-Index (sector, symbol) ---')
print(df_multi_indexed)

# --- Sorting by multiple columns ---
# Sort by 'sector' A→Z (ascending=True), then within each sector
# sort by 'market_cap' from largest to smallest (ascending=False)
df_sorted_multi = df_sample.sort_values(
    by=['sector', 'market_cap'],
    ascending=[True, False]
)
print('\n--- DataFrame sorted by sector (A-Z) and market_cap (largest first) ---')
print(df_sorted_multi)
````

## Filtering Rows with Conditions

We often need to **select only the rows that meet certain criteria**. For example:
*"Show me only Technology and Financial Services companies with a market cap above average."*

How it works:
1. A **condition** like `df['market_cap'] > 100` produces a column of `True`/`False` values (called a **boolean mask**)
2. We can combine multiple conditions with `&` (AND) or `|` (OR)
3. Passing the mask inside `df[...]` keeps only the rows where the result is `True`

````python
# First, compute the average market_cap (we need this as our threshold)
average_market_cap = df_sample['market_cap'].mean()
print(f'Average market cap: {average_market_cap:,.0f}')

# Condition 1: market_cap is above average
above_average = df_sample['market_cap'] > average_market_cap

# Condition 2: sector is either "Technology" or "Financial Services"
# isin() checks if each value is in the given list
selected_sectors = df_sample['sector'].isin(['Technology', 'Financial Services'])

# Combine both conditions with & (AND) — both must be True
condition = above_average & selected_sectors

# Apply the filter: keep only rows where condition is True
df_filtered = df_sample[condition]

print('\n--- Filtered: market_cap > average AND sector is Technology or Financial Services ---')
print(df_filtered)
````

## Operations on Text (String) Columns

Pandas provides string methods via `.str` to work with text data. In this section we will:

- Get the **unique values** in a column and count how often each appears
- **Convert** text to uppercase
- **Search** for rows containing a specific word
- **Replace** specific values with new ones
- **Map** values to a simplified version using a dictionary

````python
# --- Getting unique values ---
unique_sectors = df_sample['sector'].unique()
print('--- Unique Sectors ---')
print(unique_sectors)

# --- Counting occurrences ---
# How many companies are in each sector?
sector_counts = df_sample['sector'].value_counts()
print('\n--- Count of Each Sector ---')
print(sector_counts)

# --- Converting text to uppercase ---
# .str gives access to string methods on a whole column
df_sample['sector_upper'] = df_sample['sector'].str.upper()
print('\n--- DataFrame with uppercase sector column added ---')
print(df_sample)

# --- Searching for text inside a column ---
# Find all rows where 'sector' contains the word "tech" (case-insensitive)
df_tech = df_sample[df_sample['sector'].str.contains('tech', case=False, na=False)]
print('\n--- Rows where sector contains "tech" ---')
print(df_tech)

# --- Replacing specific values ---
# Change "Financial Services" to "Finance" everywhere in the column
df_sample['sector_replaced'] = df_sample['sector'].replace({'Financial Services': 'Finance'})
print('\n--- DataFrame with "Financial Services" replaced by "Finance" ---')
print(df_sample)

# --- Mapping values using a dictionary ---
# map() replaces values that match a key; non-matching values become NaN
sector_mapping = {
    'Technology': 'Tech',
    'Financial Services': 'Finance',
    'Industrials': 'Industrial'
}
df_sample['sector_mapped'] = df_sample['sector'].map(sector_mapping)
print('\n--- DataFrame with mapped sector names (unmatched sectors become NaN) ---')
print(df_sample)
````
